```python
import pandas as pd
 
```

```python
# 1. carregar os dados
df = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv", sep=';', encoding='utf-8', low_memory=False)
```

```python
# 2. visualizar colunas para identificar nome da delegacia e tipo de crime
print(df.columns)
# Exemplo: pode haver colunas como 'CISP', 'Delegacia', 'Ano', 'Mes', 'RouboVeiculo', 'FurtoVeiculo', etc.
```

```python
# 3. filtrar apenas ocorrências de roubo/furto de veículos (ajuste conforme o nome das colunas)
df_veiculos = df[['cisp', 'ano', 'mes', 'roubo_veiculo', 'furto_veiculos']]
```

```python
# 5. agrupar por delegacia e sumarizar (total e média mensal)
agg = df_veiculos.groupby('cisp').agg(
    total_roubo = ('roubo_veiculo', 'sum'),
    media_mensal_roubo = ('roubo_veiculo', 'mean'),
    total_furto = ('furto_veiculos', 'sum'),
    media_mensal_furto = ('furto_veiculos', 'mean'),
    meses = ('roubo_veiculo', 'count')
).reset_index()
```

```python
# 6. identificar “outliers” — por exemplo, delegacias com media muito acima da mediana
median_roubo = agg['media_mensal_roubo'].median()
threshold = median_roubo * 2  # por exemplo, 2× a mediana
outliers = agg[agg['media_mensal_roubo'] > threshold]

print("Delegacias com média de roubos > 2× mediana:", outliers)
```

```python
rmrj = [
    "Rio de Janeiro", "Niterói", "São Gonçalo", "Duque de Caxias",
    "Nova Iguaçu", "São João de Meriti", "Belford Roxo",
    "Magé", "Itaboraí", "Mesquita", "Nilópolis",
    "Queimados", "Japeri", "Maricá", "Guapimirim",
    "Tanguá"
]
```

```python
df_rmrj = df[df["regiao"].isin(rmrj)]

df
```

```python
grupo = df_rmrj.groupby("regiao").agg(
    total_furto = ("furto_veiculos", "sum"),
    media_mensal = ("furto_veiculos", "mean"),
    mediana_mensal = ("furto_veiculos", "median"),
    max_pico = ("furto_veiculos", "max")
).reset_index()

grupo
```

```python
q1 = grupo['total_furto'].quantile(0.25)
q3 = grupo['total_furto'].quantile(0.75)
iqr = q3 - q1

limite = q3 + 1.5 * iqr

outliers = grupo[grupo['total_furto'] > limite]
outliers
```
